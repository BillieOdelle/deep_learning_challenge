```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns.
df = application_df.drop(columns=['NAME','SPECIAL_CONSIDERATIONS'])
df
```

```python
# Determine the number of unique values in each column.
# unique_value = df.nunique()
# unique_value
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_types = df['APPLICATION_TYPE'].value_counts()
application_types
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = 527
application_types_to_replace = application_types[application_types <= cutoff_value].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    df['APPLICATION_TYPE'] = df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_types = df['CLASSIFICATION'].value_counts()
classification_types
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_types = classification_types[classification_types > 1]
print(classification_types)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value = 1700
classifications_to_replace = application_types[application_types <= cutoff_value].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    df['CLASSIFICATION'] = df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
df = df.drop('CLASSIFICATION', axis=1)
df_encoded = pd.get_dummies(df)
df_encoded
```

```python
# Split our preprocessed data into our features and target arrays
X = df_encoded.drop(['IS_SUCCESSFUL'], axis=1)
y = df_encoded['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=110)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the deep learning model
nn_model = tf.keras.models.Sequential()
#First Layer
nn_model.add(tf.keras.layers.Dense(units=140, activation="relu", input_dim=X_train.shape[1]))
#Second Layer
nn_model.add(tf.keras.layers.Dense(units=20, activation="relu"))
#Output
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
# Compile the Sequential model together and customise metrics
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=10)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn_model.save('AlphabetSoupCharity_Optimisation_Final.h5')
loaded_model = tf.keras.models.load_model('AlphabetSoupCharity_Optimisation_Final.h5')
```
